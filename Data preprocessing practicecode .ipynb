```python
import pandas as pd
from scipy import stats
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/sample_with_duplicates.csv')
df.head()
```

```python
df['name'].fillna('Unknown', inplace=True)
df['age']=df['age'].fillna(df['age'].mean())
df['salary']=df['salary'].fillna(df['salary'].mode()[0])
df['joining date']=pd.to_datetime(df['joining date'])
df['joining date']=df['joining date'].fillna(df['joining date'].median())
df['is Active']=df['is Active'].fillna(df['is Active'].mode()[0])
df['commnts']=df['commnts'].fillna(df['commnts'].mode()[0])
df.drop_duplicates(inplace=True)
# df.drop(["name_duplicate", "salary_duplicate"], axis=1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['name'] = le.fit_transform(df['name'])
df['commnts']=le.fit_transform(df['commnts'])

df.head()
```

```python
dfcleaned=pd.get_dummies(df,columns=['is Active'])

dfcleaned.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
dfcleaned['joining date']=scaler.fit_transform(dfcleaned[['joining date']])
dfcleaned.head()
```
