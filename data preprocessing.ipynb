```python
data = {
    'Name': ['Ali', 'Sara', 'Zain', 'Hina', 'Ahmed', 'Rida', 'Usman', 'Fatima', 'Bilal', 'Ayesha',
             'Kashif', 'Laiba', 'Tariq', 'Nida', 'Fahad', 'Mina', 'Saad', 'Anum', 'Arif', 'Sana'],

    'Age': [25, 22, None, 28, 30, None, 24, 27, None, 23,
            29, None, 31, 26, None, 21, 33, None, 32, None],

    'Salary': [50000, 55000, None, 60000, None, 52000, 58000, None, 61000, None,
               53000, 57000, None, None, 59000, 54000, None, 56000, None, 62000],

    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female', 'Male', None, 'Male', None, 'Male', None],

    'Joining Date': ['2022-01-10', '2021-05-15', None, '2020-08-20', '2023-03-12', None, '2019-11-01',
                     '2022-06-18', None, '2021-09-07', '2020-02-22', None, '2018-12-30', '2022-04-10',
                     None, '2023-01-25', '2017-07-05', None, '2016-10-19', None],

    'Comments': ['Hardworking', None, 'Needs improvement', 'Excellent team player', None, 'Quick learner',
                 'Consistent', None, 'On probation', None, 'Punctual', None, 'Senior staff', 'Good communication',
                 None, 'New joiner', 'Promoted recently', None, 'Technical expert', None],

    'Active Status': [True, True, None, True, False, True, None, True, False, None,
                      True, False, None, True, None, True, True, None, False, None]
}

df=pd.DataFrame(data)
print (df)
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Salary']=df['Salary'].fillna(df['Salary'].median())
df['Gender']=df['Gender'].fillna(df['Gender'].mode())
df['Joining Date']=df['Joining Date'].fillna(df['Joining Date'].min)
df['Comments']=df['Comments'].fillna("info not updated yet")
df['Active Status']=df['Active Status'].fillna(df['Active Status'].mode()[0])
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Active Status']=le.fit_transform(df['Active Status'])
```

```python
import pandas as pd
onehot=pd.get_dummies(df['Gender'])
df=pd.concat([df,onehot],axis=1).drop(columns=['Gender'])
df.to_csv("Cleaned",index=False  )
print("code done")
```
